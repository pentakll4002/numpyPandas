```python
import pandas as pd
```

```python
### Series
# "Pandas Series là một đối tượng giống như mảng "MỘT CHIỀU" có thể chứa bất kỳ kiểu dữ liệu nào, tương tự như một cột trong bảng."

import pandas as pd
data = [1, 2, 3, 4, 5]
series = pd.Series(data)
print("Series: \n", series)
print(type(series))
```

```python
## Create a Series from dictionary
data = {'a':1, 'b':2, 'c':3}
series_dict = pd.Series(data)
print("Series Dictionary: \n", series_dict)
```

```python
### Index: chỉ mục

data = [10, 20, 30]
index = ['a', 'b', 'c']
series = pd.Series(data, index=index)
print(series)
```

***DataFrame***

```python
## Dataframe
## Create a DataFrame from a dictionary of list

data = {
    'Name': ['Krish', 'Ân', 'Nga'],
    'Age': [25, 20, 21],
    'City': ['Banglades', 'VN', 'TQ'],
}
df = pd.DataFrame(data)
print(df)
```

ACCESS DATA FROM DATAFRAME


```python
## Dataframe
## Create a DataFrame from a dictionary of list

data = {
    'Name': ['Krish', 'Ân', 'Nga'],
    'Age': [25, 20, 21],
    'City': ['Banglades', 'VN', 'TQ'],
}
df = pd.DataFrame(data)
print(df)
```

```python
## Lấy cột Name
df['Name']
```

```python
## Chọn hàng số 0, LOC nghĩa là chỉ số hàng, sử dụng truy cập dữ liệu
df.loc[0][0]
```

```python
### ILOC là để truy cập chỉ số nguyên
df.iloc[0][0]
```

ACCESSING A SPECIFIED ELEMENT



```python
df
```

```python
### Tìm tới vị trí thứ 2 trong cột Age
print(df.at[1, 'Age'])
### Tìm tới vị trí thứ 2 trong cột Name
print(df.at[1, 'Name'])
```

ACCESSING A SPECIFIED ELEMENT BY IAT



```python
### Nếu sử dụng iloc, iat thì đây là chỉ số cột
df.iat[2, 2]
```
